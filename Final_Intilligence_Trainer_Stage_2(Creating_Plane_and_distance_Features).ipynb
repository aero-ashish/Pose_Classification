```python
from sympy import Point, Point3D, Plane, Line3D
#sympy.__version__
```

```python
n  = []
w1 = []
w2 = []
e1 = []
e2 = []
s1 = []
s2 = []
k1 = []
k2 = []
h1 = []
h2 = []



#Create a plane using 3 points with the help of SymPy Library 

for i in range(len(plank)):
  p1 = Plane(Point3D(plank['left_wrist_X'][i], plank['left_wrist_Y'][i], plank['left_wrist_Z'][i]), 
           Point3D(plank['right_wrist_X'][i], plank['right_wrist_Y'][i], plank['right_wrist_Z'][i]), 
           Point3D(plank['left_foot_X'][i], plank['left_foot_Y'][i], plank['left_foot_Z'][i]))
  
  p2 = p1.equation()

  #Distance to nose
  nose1 = Point3D(plank['nose_X'][i], plank['nose_Y'][i], plank['nose_Z'][i])
  nd1 = p1.distance(nose1)

  n.append((nd1).evalf())
  
  #Distance to wrist 1 & 2
  wr1 = Point3D(plank['left_wrist_X'][i], plank['left_wrist_Y'][i], plank['left_wrist_Z'][i])
  wd1 = p1.distance(wr1)


  wr2 = Point3D(plank['right_wrist_X'][i], plank['right_wrist_Y'][i], plank['right_wrist_Z'][i])
  wd2 = p1.distance(wr2) 
  
  w1.append((wd1).evalf())
  w2.append((wd2).evalf())

  #Distance to elbow 1 & 2
  elb1 = Point3D(plank['left_elbow_X'][i], plank['left_elbow_Y'][i], plank['left_elbow_Z'][i])
  ed1 = p1.distance(elb1)


  elb2 = Point3D(plank['right_elbow_X'][i], plank['right_elbow_Y'][i], plank['right_elbow_Z'][i])
  ed2 = p1.distance(elb2) 
  
  e1.append((ed1).evalf())
  e2.append((ed2).evalf())
  
  #Distance to shoulder 1 & 2
  sld1 = Point3D(plank['left_shoulder_X'][i], plank['left_shoulder_Y'][i], plank['left_shoulder_Z'][i])
  sd1 = p1.distance(sld1)


  sld2 = Point3D(plank['right_shoulder_X'][i], plank['right_shoulder_Y'][i], plank['right_shoulder_Z'][i])
  sd2 = p1.distance(sld2) 
  
  s1.append((sd1).evalf())
  s2.append((sd2).evalf())

  #Distance to knee 1 & 2
  knee1 = Point3D(plank['left_knee_X'][i], plank['left_knee_Y'][i], plank['left_knee_Z'][i])
  kd1 = p1.distance(knee1)


  knee2 = Point3D(plank['right_knee_X'][i], plank['right_knee_Y'][i], plank['right_knee_Z'][i])
  kd2 = p1.distance(knee2) 
  
  k1.append((kd1).evalf())
  k2.append((kd2).evalf())

  #Distance to Hip 1 & 2
  hip1 = Point3D(plank['left_hip_X'][i], plank['left_hip_Y'][i], plank['left_hip_Z'][i])
  hd1 = p1.distance(hip1)


  hip2 = Point3D(plank['left_hip_X'][i], plank['left_hip_Y'][i], plank['left_hip_Z'][i])
  hd2 = p1.distance(hip2) 
  
  h1.append((hd1).evalf())
  h2.append((hd2).evalf())

  

#Creating columns in dataframe
plank['nose_dist']        = n
#plank['wrist_dist_1']     = w1
#plank['wrist_dist_2']     = w2
plank['elbow_dist_1']     = e1
plank['elbow_dist_2']     = e2
plank['shoulder_dist_1']  = s1
plank['shoulder_dist_2']  = s2
plank['knee_dist_1']      = k1
plank['knee_dist_2']      = k2
plank['hip_dist_1']       = h1
plank['hip_dist_2']       = h2
```

```python
angl1 = []
angr1 = []
angl2 = []
angr2 = []
angl3 = []
angr3 = []
angl4 = []
angr4 = []

#Create Angles with the help of SymPy Library 

for i in range(len(plank)):
  rst1   = Point3D(plank['left_wrist_X'][i], plank['left_wrist_Y'][i], plank['left_wrist_Z'][i])
  rst2   = Point3D(plank['right_wrist_X'][i], plank['right_wrist_Y'][i], plank['right_wrist_Z'][i])
  elb1   = Point3D(plank['left_elbow_X'][i], plank['left_elbow_Y'][i], plank['left_elbow_Z'][i])
  elb2   = Point3D(plank['right_elbow_X'][i], plank['right_elbow_Y'][i], plank['right_elbow_Z'][i])
  sld1   = Point3D(plank['left_shoulder_X'][i], plank['left_shoulder_Y'][i], plank['left_shoulder_Z'][i])
  sld2   = Point3D(plank['right_shoulder_X'][i], plank['right_shoulder_Y'][i], plank['right_shoulder_Z'][i])
  hip1   = Point3D(plank['left_hip_X'][i], plank['left_hip_Y'][i], plank['left_hip_Z'][i])
  hip2   = Point3D(plank['right_hip_X'][i], plank['right_hip_Y'][i], plank['right_hip_Z'][i])
  knee1  = Point3D(plank['left_knee_X'][i], plank['left_knee_Y'][i], plank['left_knee_Z'][i])
  knee2  = Point3D(plank['right_knee_X'][i], plank['right_knee_Y'][i], plank['right_knee_Z'][i])
  ankle1 = Point3D(plank['left_ankle_X'][i], plank['left_ankle_X'][i], plank['left_ankle_X'][i]) 
  ankle2 = Point3D(plank['right_ankle_X'][i], plank['right_ankle_Y'][i], plank['right_ankle_Z'][i])
  
  lh1, rh1 = Line3D(rst1, elb1), Line3D(rst2, elb2)
  lh2, rh2 = Line3D(elb1, sld1), Line3D(elb2, sld2)
  lh3, rh3 = Line3D(sld1, hip1), Line3D(sld2, hip2)
  lh4, rh4 = Line3D(hip1, knee1), Line3D(hip2, knee2)
  lh5, rh5 = Line3D(knee1, ankle1), Line3D(knee2, ankle2)
  
  #Angle belween wrist, elbow & shoulder
  al1 = lh1.smallest_angle_between(lh2)  
  ar1 = rh1.smallest_angle_between(rh2)
  
  angl1.append((al1).evalf())
  angr1.append((ar1).evalf())

  #Angle belween elbow, shoulder & hip
  al2 = lh2.smallest_angle_between(lh3)  
  ar2 = rh2.smallest_angle_between(rh3)
  
  angl2.append((al2).evalf())
  angr2.append((ar2).evalf())

  #Angle belween shoulder, hip & knee
  al3 = lh3.smallest_angle_between(lh4)  
  ar3 = rh3.smallest_angle_between(rh4)

  angl3.append((al3).evalf())
  angr3.append((ar3).evalf())
  
  #Angle belween hip, knee & ankle
  al4 = lh4.smallest_angle_between(lh5)  
  ar4 = rh4.smallest_angle_between(rh5)

  angl4.append((al4).evalf())
  angr4.append((ar4).evalf())

plank['wrist_elb_sld_L']  = angl1
plank['wrist_elb_sld_R']  = angr1                                  
plank['elb_sld_hip_L']    = angl2
plank['elb_sld_hip_R']    = angr2
plank['sld_hip_knee_L']   = angl3
plank['sld_hip_knee_R']   = angr3
plank['hip_knee_ankle_L'] = angl4
plank['hip_knee_ankle_R'] = angr4
```
